# Бенчмарк асинхронных запросов
Этот ноутбук демонстрирует тестирование производительности асинхронных HTTP-запросов к API WAX.

#

```python
import asyncio
import logging
import time
from eosapi import EosApi

logging.basicConfig(level=logging.INFO)
log = logging.getLogger(__name__)

#
```

```python
class PoolServer:
    def __init__(self):
        self.rpc_host = "https://wax.alienworlds.io"
        self.wax_api = EosApi(
            rpc_host=self.rpc_host,
            proxy=("65.21.182.63", 13001, 300),
            # yeomen_proxy=("65.21.182.63", 13001, 300),
        )

    async def get_planet_pools_data(self, planet_name):
        payload = {
            "json": True,
            "code": "m.federation",
            "scope": f"{planet_name}.world",
            "table": "pools",
            "key_type": "",
            "index_position": 1,
            "lower_bound": "",
            "upper_bound": "",
            "limit": 1,
        }
        resp = await self.wax_api.get_table_rows_async(payload)
        return resp

async def run_benchmark(num_requests=25):
    pool_server = PoolServer()
    start_time = time.time()

    # Последовательное выполнение запросов
    times = []
    for i in range(num_requests):
        request_start = time.time()
        result = await pool_server.get_planet_pools_data("neri")
        request_end = time.time()
        request_time = request_end - request_start
        times.append(request_time)
        print(f"Запрос {i+1}: {request_time:.2f} сек")

    end_time = time.time()
    total_time = end_time - start_time
    avg_time = total_time / num_requests

    print(f"\nСтатистика:")
    print(f"Всего запросов: {num_requests}")
    print(f"Общее время выполнения: {total_time:.2f} секунд")
    print(f"Среднее время на запрос: {avg_time:.2f} секунд")
    print(f"Минимальное время запроса: {min(times):.2f} секунд")
    print(f"Максимальное время запроса: {max(times):.2f} секунд")
    print(f"Запросов в секунду: {num_requests/total_time:.2f}")
```

```python
await run_benchmark(25)
```
